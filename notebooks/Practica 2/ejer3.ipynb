```python
#solicitar palabras al usuario
words = input('Ingrese al menos 3 palabras consideradas spoiler separadas por una coma: ')
```

```python
spoiler_words =[w.strip() for w in words.split(',')]
```

```python
if(len(spoiler_words) >= 3) :
    review_list = ['*' * len(p) if p in spoiler_words else p for p in review_list]
```
