```python
def email_valido(mail):
    value = False
    if (mail.count('@') == 1) and (mail.split('@')[0] != "") and ('.' in mail.split('@')[1]) and (mail[0] != '.' or mail[-1] != '.') and (len(mail.split('.')[1]) > 2):
        value = True
    return value
```

```python
def validar_email(mail):
    if (email_valido(mail)):
        print('su email es valido')
    else: print('su email no es valido')
```

```python
mail = input('Ingrese su mail ')
validar_email(mail)
```
