```python
def to_seconds(duration):
    sides = duration.split(':')
    return int(sides[0])*60 + int(sides[1])
```

```python
#calcular el total de la playlist
def total_time():
    total_seconds = 0
    for cancion in playlist:
        total_seconds += to_seconds(cancion["duration"])
    
    total_minutes = total_seconds // 60
    total_seconds = total_seconds % 60
    print(f'La duracion de la playlist es de {total_minutes}:{total_seconds}')
```

```python
def longest_song():
    return max(playlist, key=lambda c: to_seconds(c["duration"]))

def shortest_song():
    return min(playlist, key=lambda c: to_seconds(c["duration"]))
```

```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
```

```python
longest = longest_song()
shortest = shortest_song()
print(f'la cancion mas larga es "{longest["title"]}" ({longest["duration"]})')
print(f'la cancion mas corta es "{shortest["title"]}" ({shortest["duration"]})');
```
