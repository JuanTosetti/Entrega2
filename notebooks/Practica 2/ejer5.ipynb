```python
rates = {
    "local":[500,1000,2000],
    "regional":[1000,2500,5000],
    "nacional":[2000,4500,8000]
}
```

```python
def calcular_indice(weight):
    if(weight < 1):
        return 0
    elif(1 < weight <= 5):
        return 1
    else: return 2
```

```python
def calcular_costo(z, w):
    indice = calcular_indice(w)
    costo = rates[z][indice]
    print(f'el costo del envio {zone} es de ${costo}')
```

```python
#Entrada de datos
def calcular_costo_de_envio():
    weight = float(input('Ingresa el peso en kilos: '))
    zone = input('Ingresa la zona de envios (local, regional, nacional): ').strip().lower()

    if zone not in rates:
        print('Error zona no valida ');
    else:
        calcular_costo(zone, weight)
```

```python
calcular_costo_de_envio()
calcular_costo_de_envio()
```
