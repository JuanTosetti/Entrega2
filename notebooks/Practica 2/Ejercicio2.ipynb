```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]

durations = [song['duration'] for song in playlist]
durations
```

```python
#Ahora ordenamos
durations.sort()
durations
```

```python
seconds = []
minutes = []
seconds_final = 0
for elem in durations:
    s = elem.split(':')[1]
    seconds.append(int(s))
    m =elem.split(':')[0]
    minutes.append(int(m))
    
seconds_final = sum(seconds) 
minutes = sum(minutes) + (seconds_final // 60)
seconds_final = seconds_final % 60
print(f'Total de minutos: {minutes}:{seconds_final}')
```
